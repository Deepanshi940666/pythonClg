```python
def evenFact(n):
    fact=1
    for i in range(2,n+1,2):
        fact=fact*i
    return fact

def oddFact(n):
    fact=1
    for i in range(1,n+1,2):
        fact=fact*i
    return fact
# fact=1
n=int(input("Enter a no:"))
if(n%2==0):
    print(evenFact(n))
else:
    print(oddFact(n))


```

```python
#calculate nCr
def fact(x):
    fact=1
    for i in range(1,x+1):
        fact*=i
    return fact

n=10
r=2
nfact=fact(n)
rfact=fact(r)
nrfact=fact(n-r)
ans= nfact/(rfact*nrfact)
print(ans)
```
